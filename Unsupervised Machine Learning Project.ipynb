# Import data

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
sns.set()
```

```python
#Load the data
data = pd.read_csv('iris_dataset.csv')
data
```

```python
#Plot the data
plt.scatter(data['sepal_length'], data['sepal_width'])
plt.xlabel('Length of sepal')
plt.ylabel('Width of sepal')
plt.show()
```

## Clustering unscaled data

```python
x = data.copy()
kmeans = KMeans(2)
kmeans.fit(x)
```

```python
clusters = data.copy()
clusters['pred'] = kmeans.fit_predict(x)
```

```python
plt.scatter(data['sepal_length'], data['sepal_width'], c = clusters['pred'], cmap = 'rainbow')
```

## Standardize the variables

```python
from sklearn import preprocessing
```

```python
x_scaled = preprocessing.scale(x)
x_scaled
```

## Clustering standardized data

```python
kmeans_scaled = KMeans(2)
kmeans_scaled.fit(x_scaled)
```

```python
clusters_scaled = data.copy()
clusters_scaled['cluster_pred'] = kmeans_scaled.fit_predict(x_scaled)
clusters_scaled
```

```python
#plot
plt.scatter(clusters_scaled['sepal_length'], clusters_scaled['sepal_width'], c = clusters_scaled['cluster_pred'], cmap = 'rainbow')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal width')
plt.show()
```

## Identifying cluster number using Elbow Method

```python
#WCSS
wcss = []
cl_num
for i in range(1,cl_num):
    kmeans = KMeans(i)
    kmeans.fit(x_scaled)
    wcss.append(kmeans.inertia_)
wcss
```

```python
number_of_clusters= range(1,cl_num)
plt.plot(number_of_clusters,wcss)
plt.title('Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('With-Cluster Sum of Squares')
```
